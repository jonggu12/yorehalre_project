### 1.래시피 링크 가져오기(총 869개)

```python
import requests
from bs4 import BeautifulSoup
import time

recipe_name_list = []
recipe_image_list = []
recipe_link_list = []
code_list = []

for code in theme_code_df["theme_code"]:
    link = f"https://www.10000recipe.com/theme/view.html?theme={code}&page="
    for page in range(1, 6):
        url = link + str(page)
        response = requests.get(url)
        rescode = response.status_code
        if(rescode==200):
            html = response.text
            soup = BeautifulSoup(html, "html.parser")
            content = soup.find_all(class_="common_sp_list_li")
            for item in content:
                recipe_name = item.find("div", class_="common_sp_caption_tit line2").text.strip()
                recipe_link = item.find("a", class_="common_sp_link")["href"]
                recipe_image_link = item.find("img")["src"]
                recipe_name_list.append(recipe_name)
                recipe_link_list.append("https://www.10000recipe.com/" + recipe_link)
                recipe_image_list.append(recipe_image_link)
                code_list.append(code)
        else:
            print("Page not found")
    time.sleep(1)

print(len(recipe_name_list))
print(len(recipe_link_list))
print(len(recipe_image_list))


```

```python
df =  pd.DataFrame({"theme_code":code_list, 
                    "Recipe Name": recipe_name_list, 
                    "Recipe Link":recipe_link_list, 
                    "Recipe Image Link": recipe_image_list})
df.head()
```

```python
theme_code_df['theme_code'] = theme_code_df['theme_code'].astype('int64')

concat_df = pd.merge(df, theme_code_df, on="theme_code",how="left")
concat_df.head()
```

```python
concat_df.to_csv("./recipe_links.csv",index=False)
```

### 2.레시피 요약 재료 및 소리순서 가져오기 

```python
def get_recipe_detail(soup):
    content = soup.find("div", class_="view_step")
    try: 
        summary = soup.find("div", class_="view2_summary_in").text.strip()
    except:
        summary = "NaN"
    
    try:           
        ingredients = soup.find("div", class_="ready_ingre3")
        ingredient_name = ingredients.get_text(separator=' ', strip=True).replace("구매", "")
    except:
        ingredient_name = "NaN"
        
    steps = content.find_all("div", class_="media-body")
    full_step = []
    for step in steps:
        text = step.text.strip()
        full_step.append(text)
    final_step = " ".join(full_step)
    
    return summary, ingredient_name, final_step
```

```python
import time 
for index, link in enumerate(df["Recipe Link"]):
    response = requests.get(link)
    soup = BeautifulSoup(response.text, "html.parser")
    summary, ingredient_name, final_step = get_recipe_detail(soup)
    df.at[index, "summary"] = summary
    df.at[index, "ingredient_name"] = ingredient_name
    df.at[index, "steps"] = final_step
    time.sleep(2)

df.head()
```
