# 연도별 인구현황

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel('.\data\성_및_연령별_추계인구_1세별__5세별____전국_20240228145407.xlsx')
```

```python
data.drop(columns='가정별', axis = 1, inplace = True)
```

```python
data.drop(range(23, len(data)), inplace=True)
```

```python
data.drop(columns='성별', axis = 1, inplace = True)
```

```python
data = data.replace("-", 0)
data = data.set_index('연령별')
```

```python
data.index
```

```python
people_1984 = data.loc['0 - 4세':,'1984']
people_2024 = data.loc['0 - 4세':,'2024']
people_2064 = data.loc['0 - 4세':,'2064']
```

```python
people_2024.drop(index='80세이상', inplace = True)
people_2064.drop(index='80세이상', inplace = True)
```

```python
people_1984.index
```

# 서울 시민의 소비 및 이동 패턴

```python
import pandas as pd

data = pd.read_csv('./data/서울 시민의 소비 및 이동 패턴 재현 데이터.csv')
list = []
for a in data.columns:
    list.append(a)
print(list)
```

```python
select_column = [col for col in data.columns if '가맹점분류' in col]
select_column.insert(0, '생년월일')
select_column
data1 = data[select_column]
data1 = data1.dropna()
```

```python
from datetime import datetime
Y_year = datetime.now().year - 60
Y = data1.loc[data1['생년월일']>Y_year]
O = data1.loc[data1['생년월일']<=Y_year]
print(Y.info(), O.info())
```

```python
Y.describe()
Y_mean_dict = Y.mean().to_dict()
del Y_mean_dict['생년월일']
```

```python
O.describe()
O.mean()
O_mean_dict = O.mean().to_dict()
del O_mean_dict['생년월일']
O_mean_dict
```

```python
Y_mean = pd.DataFrame(Y_mean_dict, index=[0])
O_mean = pd.DataFrame(O_mean_dict, index=[0])
```

```python
total_mean = pd.concat([Y_mean, O_mean])
total_mean
```

```python
total_mean.rename(columns={'가맹점분류_이용비중1':'요식', '가맹점분류_이용비중2':'유흥', '가맹점분류_이용비중3':'백화점', '가맹점분류_이용비중4':'대형할인점','가맹점분류_이용비중5':'소형유통점',
                           '가맹점분류_이용비중6':'의류/잡화', '가맹점분류_이용비중7':'의류/잡화','가맹점분류_이용비중8':'의류/잡화','가맹점분류_이용비중9':'여행',
                           '가맹점분류_이용비중10':'교통','가맹점분류_이용비중11':'미용','가맹점분류_이용비중12':'가정생활','가맹점분류_이용비중13':'교육',
                           '가맹점분류_이용비중14':'의료','가맹점분류_이용비중15':'가전/가구','가맹점분류_이용비중16':'자동차판매','가맹점분류_이용비중17':'자동차서비스',
                           '가맹점분류_이용비중18':'주유','가맹점분류_이용비중19':'전자상거래'}, inplace=True)
```
